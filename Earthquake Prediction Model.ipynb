```python
usgs_df = usgs_df[['time', 'latitude', 'longitude', 'depth', 'mag']]
```

```python
usgs_df.head()
```

```python
usgs_df['time'] = pd.to_datetime(usgs_df['time'])
```

```python
usgs_df['year'] = usgs_df['time'].dt.year
usgs_df['month'] = usgs_df['time'].dt.month
usgs_df['day'] = usgs_df['time'].dt.day
```

```python
usgs_df.head()
```

```python
y_kaggle = kaggle_df['magnitude']
```

```python
X_kaggle = kaggle_df.drop(['magnitude', 'label'], axis=1)
```

```python
X_kaggle = pd.get_dummies(X_kaggle)
```

```python
X_kaggle.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

X_train, X_test, y_train, y_test = train_test_split(
    X_kaggle, y_kaggle, test_size=0.2, random_state=42
)

model = RandomForestRegressor()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

mae = mean_absolute_error(y_test, predictions)

print("Model trained successfully")
print("MAE:", mae)
```

```python
# Step 1: Prepare USGS features
X_usgs = usgs_df[['latitude', 'longitude', 'depth', 'year', 'month', 'day']]

# Step 2: Match Kaggle columns (very important)
X_usgs = X_usgs.reindex(columns=X_kaggle.columns, fill_value=0)

# Step 3: Predict
usgs_predictions = model.predict(X_usgs)

# Step 4: Add predictions to dataset
usgs_df['predicted_magnitude'] = usgs_predictions

# Step 5: View results
usgs_df[['latitude', 'longitude', 'mag', 'predicted_magnitude']].head()
```

```python
usgs_df = usgs_df.dropna(subset=['mag', 'predicted_magnitude'])
```

```python
from sklearn.metrics import mean_absolute_error

mae_usgs = mean_absolute_error(usgs_df['mag'], usgs_df['predicted_magnitude'])

print("USGS MAE:", mae_usgs)
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.scatter(usgs_df['mag'], usgs_df['predicted_magnitude'])

plt.xlabel("Actual Magnitude")
plt.ylabel("Predicted Magnitude")
plt.title("Actual vs Predicted Earthquake Magnitude")

plt.show()
```

```python
errors = usgs_df['mag'] - usgs_df['predicted_magnitude']

plt.figure()

plt.hist(errors)

plt.xlabel("Error")
plt.ylabel("Frequency")
plt.title("Error Distribution")

plt.show()
```
